1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
x = sp.symbols('x')
```

```python
f_x=10*x - 5
```

```python
df_x = sp.diff(f_x,x)
df_x
```

```python

x_val = np.linspace(0, 10)
f = 10*x_val - 5

plt.figure(figsize=(10, 5))
plt.plot(x_val, f, label='f(x) = 10*x - 5', color='blue')
plt.axhline(10, color='green', linestyle='--', label="f'(x) = 10")

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
f_sin_x = sp.sin(x)
df_sin_x = sp.diff(f_sin_x,x)
df_sin_x
```

```python
x_val = np.linspace(0, 10)
f = np.sin(x_val)
df = np.cos(x_val)

plt.figure(figsize=(10, 5))
plt.plot(x_val, f, label='f(x) = sin(x)', color='blue')
plt.plot(x_val, df, label='f(x) = cos(x)', color='green')

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
f_ln_x = sp.ln(x)
df_ln_x = sp.diff(f_ln_x,x)
df_ln_x
```

```python
x_val = np.linspace(0.1, 10)
f = np.log(x_val)
df = 1/x_val

plt.figure(figsize=(10, 5))
plt.plot(x_val, f, label='f(x) = ln(x)', color='blue')
plt.plot(x_val, df, label='f(x) = 1/x', color='green')

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
f_x = x**2 - 5*x + 1
```

```python
df_x = sp.diff(f_x,x)
df_x
```

```python
x_val = np.linspace(0, 100)
f = x_val**2 - 5*x_val + 1
df = 2*x_val - 5

plt.figure(figsize=(10, 5))
plt.plot(x_val, f, label='f(x) = x**2 - 5*x + 1', color='blue')
plt.plot(x_val, df, label='f(x) = 2*x - 5', color='green')

plt.title("Витрати компанії")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```
